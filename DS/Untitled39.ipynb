```python
data=pd.read_csv('/content/pre_finel2.csv')
```

```python
data['Model']=data['Model'].str.replace('-' ,' ')
```

```python
lists=data.to_dict('records')
```

```python
opt=data[data['Launch_Date'].isnull()]
```

```python
opt=opt.drop_duplicates(subset='Model')
```

```python
lists=opt.to_dict('records')
```

```python
for x in tqdm(lists):
  headers = {'User-Agent':  random.choice(user_agent_list)}
  name=x['Model'].replace(' ','+')
  url=f'https://www.cpu-upgrade.com/cgi-bin/SearchSite.pl?SEARCH=amd+{name}'
  response=requests.get(url,headers=headers)
  soup=BeautifulSoup(response.content,'html.parser')
  bucket=soup.find_all('div',class_='ss_results')
  print(url)
  if len(bucket) >0:
    break
```
